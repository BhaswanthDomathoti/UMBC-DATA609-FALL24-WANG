## In-class Exercise

Create a new column call "GDP Quartile" in the df_2020 dataframe and assign each country the quartile it belongs to by applying the assign_quartile() function.


```python
# Define the DataFrame (assuming df_2020 is already loaded with a 'GDP' column)
# df_2020 = pd.read_csv("path_to_your_csv.csv")

def assign_quartile(row, quartiles):
    if row['GDP per Capita'] <= quartiles[0.25]:
        return 'Q1'
    elif row['GDP per Capita'] <= quartiles[0.5]:
        return 'Q2'
    elif row['GDP per Capita'] <= quartiles[0.75]:
        return 'Q3'
    else:
        return 'Q4'

# Calculate quartiles for the 'GDP' column
quartiles = df_2020['GDP per Capita'].quantile([0.25, 0.5, 0.75])

# Apply the assign_quartile function to each row in df_2020
df_2020['GDP Quartile'] = df_2020.apply(assign_quartile, axis=1, quartiles=quartiles)

# Check the result
print(df_2020[['Country', 'GDP per Capita', 'GDP Quartile']].head())

```
